```python
import random
import csv
filename='database.csv'

def BuyerLogin():
    print("BUYER LOGIN PORTAL")
    username=input('Enter userid:')
    password=input('Enter password:')
    with open(filename,'r') as file:
        data=csv.reader(file)
        datals=list(data)
        for row in datals:
            if username==row[0] and password==row[2] and row[4]=='buyer':
                BuyerMenu()
def BuyerMenu():
    print('-'*48)
    print("BuyerMenu")
    print("0.Exit")
    print("1.search")
    option=eval(input("give your choice to proceed further:"))
    print('#'*48)
    if option==0:
        menu()
    elif option==1:
        SearchComics()
    else:
        pass
def SearchComics():
    print('search comics\n','-'*48)
    print('-'*48)
    print("BuyerMenu")
    print("0.search by comicInfo")
    print("1.search by Language")
    print("2.search by Category")
    option=eval(input("give your choice to proceed further:"))
    print('#'*48)
    ComicInfo,comicLanguage,comicCategory='','',''
    if option==0:
        ComicInfo=input('Enter comic info:')
    elif option==1:
        comicLanguage=input('Enter comic language:')
    elif option==2:
        comicCategory=input('Enter comic category:')
    with open('comic.csv','r') as file:
        csvread=csv.reader(file)
        ls=list(csvread)
        result=[i for i in ls if ComicInfo in i or comicLanguage in i or comicCategory in i]
        if len(result)>0:
            for i in list(result):
                print(f'{i[0]}{i[1]}{i[2]},{i[3]},{i[4]},{i[5]},{i[6]},{i[7]}')
    print('Comic uploaded')
def SellerLogin():
    print("SELLER LOGIN PORTAL")
    username=input('Enter userid:')
    password=input('Enter password:')
    with open(filename,'r') as file:
        data=csv.reader(file)
        datals=list(data)
        for row in datals:
            if username==row[0] and password==row[2] and row[4]=='seller':
                SellerMenu()
def SellerMenu():
    print('-'*48)
    print("SellerMenu")
    print("0.Exit")
    print("1.upload")
    option=eval(input("give your choice to proceed further:"))
    print('#'*48)
    if option==0:
        menu()
    elif option==1:
        UploadComics()
    else:
        pass
def UploadComics():
    print('-'*48)
    print('upload comics')
    print('-'*48)
    ComicInfo=input('Enter the information of the comic:')
    ComicPrice=input('Enter the comic Price:')
    comicAuthor=input('Enter the author name:')
    comicDescription=input('Enter a short description:')
    comicStoryline=input('Enter comic storyline:')
    comicContent=input('Enter comic content:')
    comicLanguage=input('Enter comic language:')
    comicCategory=input('Enter category:')
    with open('comic.csv','a',newline='') as file:
        csvwrite=csv.writer(file)
        csvwrite.writerow([ComicInfo,ComicPrice,comicAuthor,comicDescription,comicStoryline,comicContent,comicLanguage,comicCategory])
    print('Comic uploaded')
def CreateUser():
    userid=random.randint(0,1000)
    name=input("Enter name:")
    password=input("Enter password:")
    language=input("Enter language:")
    roll=input('enter buyer/seller:')
    with open(filename,'a',newline='') as file:
        csvwriter = csv.writer(file)
        csvwriter.writerow([userid,name,password,language,roll])
        print("user created")
    print(f"Your Id is:{userid}")
def menu():
    print('-'*48)
    print("Menu")
    print("0.Exit")
    print("1.Buyer")
    print("2.Seller")
    print("3.New User")

    option=eval(input("give your choice to proceed further:"))
    print('-'*48)
    return option
while True:

    option=menu()
    if option==0:
        break
    elif option==1:
        BuyerLogin()
    elif option==2:
        SellerLogin()
    elif option==3:
        CreateUser()
    else:
        pass
```
