```python
import logging
import csv
import random
import pandas
print("welcome to the matrimony system")

filename='data.csv'
def AddDetails():
    rid=random.randint(0,1000)
    name=input("Enter name:")
    age=input("Enter age:")
    occupation=input("Enter occupation:")
    gender=input("Enter gender:")
    
    with open(filename,'a',newline='') as file:
        csvwriter = csv.writer(file)
        csvwriter.writerow([rid,name,age,occupation,gender])
        print('-'*48)
        print("Details added")
    print(f"Your Id is:{rid}")
def UpdateDetails():
    rid=int(input("Enter your id to proceed:"))
    modify=input("Enter a field to modify")
    if modify=='name':
        col=1
    elif modify=='age':
        col=2
    elif modify=='occupation':
        col=3
    elif modify=='gender':
        col=4
    with open(filename,'r') as file:
        data=csv.reader(file)
        datals=list(data)
        finalls=[]
        finalls1=[]
        for details in datals:
            print(details)
            if rid==details:
                finalls.append(details)
            else:
                finalls1.append(details)
            print(col)
        finalls[col].replace(input('enter modification'))
    with open(filename,'w') as file:
        csvwriter = csv.writer(file)
        csvwriter.writerows(finalls1)
        csvwriter.writerow(finalls)
def ShowDetails():
    print('-'*48)
    with open(filename,'r') as file:
        data=csv.reader(file)
        datals=list(data)
        if len(datals)>0:
            for row in datals:
                print(f'{row[0]}\t{row[1]}\t{row[2]}\t{row[3]}\t{row[4]}')
        else:
            print("Insert some values to display")

def SearchDetails():
    print('-'*48)
    print('search details')
    search=input("Enter a name to search:")
    with open(filename,'r') as file:
        data=csv.reader(file)
        datals=list(data)
        result=[]
        for names in datals:
            if search in names:
                result.append(names)
        print(result)
def recommendation():
    print('-'*48)

    print('recommendation')
    
def DeleteProfile():
    print('-'*48)
    name=input('Enter a name to delete')
    with open(filename,'r') as file:
        data=csv.reader(file)
        datals=list(data)
        finalls=[]
        for details in datals:
            if name!=details:
                finalls.append(details)
    with open(filename,'w') as file:
        csvwriter = csv.writer(file)
        csvwriter.writerows(finalls)
    print('Delete profile')
def menu():
    print('-'*48)
    print("Menu")
    print("0.Exit")
    print("1.Insertion")
    print("2.Modification")
    print("3.Retrieval")
    print("4.Searching")
    print("5.Recommendation")
    print("6.Delete Profile")

    option=eval(input("give your choice to proceed further:"))
    print('#'*48)
    return option
while True:

    option=menu()
    if option==0:
        break
    elif option==1:
        AddDetails()
    elif option==2:
        UpdateDetails()
    elif option==3:
        ShowDetails()
    elif option==4:
        SearchDetails()
    elif option==5:
        Recommendation()
    elif option==6:
        DeleteProfile()
    else:
        pass
    
    
```
